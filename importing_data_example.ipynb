```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/species_brainmass_neurons.csv')

mass = df.cortex_mass_g
neurons = df.Neurons

plt.scatter(mass,neurons)
plt.xscale('log')
plt.yscale('log')
```

Now lets subset the data to separate primates and non primates

```python
primate_df = df[df['Order'] == 'Primata']
nonprimate_df = df[df['Order'] != 'Primata']

p_mass = primate_df.cortex_mass_g
p_neurons = primate_df.Neurons

np_mass = nonprimate_df.cortex_mass_g
np_neurons = nonprimate_df.Neurons

plt.scatter(p_mass,p_neurons)
plt.scatter(np_mass,np_neurons)

plt.xscale('log')
plt.yscale('log')
```

```python
fig, ax = plt.subplots(1,2,figsize=(10,5))

ax[0].scatter(p_mass,p_neurons)
ax[1].scatter(np_mass,np_neurons)

plt.xscale('log')
plt.yscale('log')
```
